# Exercise 1
- Using A = A.T @ A will always generate positive eigenvalues, so there are never mismatches
- However, using matrices that are symmetric but not positive definite can yield sign mismatches between the eigenvalues
- Regardless of matrix definiteness, there can be sign mismatches between the eigenvector outputs
- Notice that for square matrices, $U = V$

```python
A = np.random.randn(5,5)
A = A.T @ A

# evals/evecs sorted by eigenvalue
evals, evecs = np.linalg.eig(A)
sidx = np.argsort(evals)[::-1]
evals, evecs = evals[sidx], evecs[:,sidx]

# SVD
U, S, Vt = np.linalg.svd(A)

print("Eigenvalues and singular values:")
print(np.vstack((evals,S)))
print("\nLeft and right singular vectors:")
print(np.round(U - Vt.T, 10))
print("\nSingular vectors minus eigenvectors (should be zero)")
print(np.round(U - evecs, 10), "\n")
print(np.round(U + evecs, 10))
```

Because U and Vt are orthogonal matrices, they should not affect matrix norms

```python
U, S, Vt = np.linalg.svd(np.random.randn(5,5))
w = np.random.randn(5,1)

print(np.linalg.norm(w))
print(np.linalg.norm(U @ w))
print(np.linalg.norm(Vt @ w))
```

# Exercise 4
- To generate a matrix with a given condition number, need to generate singular values where $\sigma_{max}/\sigma_{min} = C$
- Can then use QR decomposition to generate random orthogonal matices to multiply together

```python
m = 10
n = 6
condnum = 42

s = np.linspace(condnum, 1, np.min((m,n)))
S = np.zeros((m,n))
np.fill_diagonal(S,s)

U, _ = np.linalg.qr(np.random.randn(m,m))
Vt, _ = np.linalg.qr(np.random.randn(n,n))

A = U @ S @ Vt

fig, ax = plt.subplots(1, 4, figsize = (12,6))
ax[0].imshow(A, aspect = "equal", cmap = "gray")
ax[0].set_title(f"A (cond = {np.linalg.cond(A):.3f})")
ax[1].imshow(U, aspect = "equal", cmap = "gray")
ax[1].set_title(f"U (cond = {np.linalg.cond(U):.3f})")
ax[2].imshow(S, aspect = "equal", cmap = "gray")
ax[2].set_title(f"S (cond = {np.linalg.cond(S):.3f})")
ax[3].imshow(Vt, aspect = "equal", cmap = "gray")
ax[3].set_title(f"Vt (cond = {np.linalg.cond(Vt):.3f})")
plt.tight_layout()
plt.savefig('Figure_14_04.png',dpi=300)
plt.show()
```

# Exercise 5
- This problem starts by making a grid of the 6 standard deviations for a bivariate normal
- Each grid location shows the likelihood of seeing an observation that extreme
- We then generate a random matrix and perform a convolution with the bivariate normal kernel
- In other words, we smooth the random matrix A by judging how likely a the surrounding grid of values is at each point

```python
m = 40
n = 30

k = (m + n) // 4
X, Y = np.meshgrid(np.linspace(-3,3,k), np.linspace(-3,3,k))
g2d = np.exp(-(X**2 + Y**2) / (k/8))

from scipy.signal import convolve2d
A = convolve2d(np.random.randn(m,n), g2d, mode = "same")

U, s, Vt = np.linalg.svd(A)
S = np.zeros(np.shape(A))
np.fill_diagonal(S, s)

# Decomposition plots
fig, ax = plt.subplots(1, 4, figsize = (12,6))
ax[0].imshow(A, aspect = "equal", cmap = "gray", vmin = -10, vmax = 10)
ax[0].set_title("A")
ax[1].imshow(U, aspect = "equal", cmap = "gray")
ax[1].set_title("U")
ax[2].imshow(S, aspect = "equal", cmap = "gray")
ax[2].set_title("S")
ax[3].imshow(Vt, aspect = "equal", cmap = "gray")
ax[3].set_title("Vt")
plt.tight_layout()
plt.savefig('Figure_14_05a.png',dpi=300)
plt.show()

# Scree plot
plt.figure(figsize = (12,3))
plt.plot(100 * s / np.sum(s), "ks-", markersize = 10)
plt.xlabel("Component index")
plt.ylabel("Variance explained (%)")
plt.title("Scree plot")
plt.savefig('Figure_14_05b.png',dpi=300)
plt.show()
```

```python
layers = 4
rank1mats = np.zeros((layers, m, n))
fig, ax = plt.subplots(2, layers, figsize = (10,6))

for i in range(layers):
    rank1mats[i,:,:] = np.outer(U[:,i], Vt[i,:]) * S[i,i]
    ax[0,i].imshow(rank1mats[i,:,:], cmap = "gray", vmin = -10, vmax = 10)
    ax[0,i].set_title(f"Layer {i}")
    ax[0,i].set_xticks([])
    ax[0,i].set_yticks([])
    
    ax[1,i].imshow(np.sum(rank1mats[:i+1,:,:], axis = 0), cmap = "gray", vmin = -10, vmax = 10)
    ax[1,i].set_title(f"Layer 0:{i}")
    ax[1,i].set_xticks([])
    ax[1,i].set_yticks([])
    
plt.tight_layout()
plt.savefig('Figure_14_05c.png',dpi=300)
plt.show()
```

# Exercise 6
- To form MP pseudoinverse, need to invert nonzero values of $\Sigma$ with a reasonable margin of error
- To do this, can use the numpy float precision and scale it by the size of the array

```python
A = np.random.randn(5,3) @ np.random.randn(3,5)
U, s, Vt = np.linalg.svd(A)
tol = np.finfo(float).eps * np.max(A.shape)

sinv = np.zeros_like(s)
sinv[s > tol] = 1/s[s > tol]

S = np.zeros_like(A)
np.fill_diagonal(S, sinv)
Apinv = Vt.T @ S @ U.T

ApinvNP = np.linalg.pinv(A)
print(np.round(ApinvNP - Apinv, 5))
```
