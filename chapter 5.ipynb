# Exercise 2

```python
original = np.arange(100).reshape(10,10)
C = original[:5,:5].copy()

fig, ax = plt.subplots(1, 2, figsize = (10,5))

# Plot original matrix, draw dashed lines, plot numbers for each cell
ax[0].imshow(original, cmap = "gray", origin = "upper", vmin = 0, vmax = np.max(original))
ax[0].plot([4.5, 4.5], [-0.5, 9.5], "w--")
ax[0].plot([-0.5, 9.5], [4.5, 4.5], "w--")
ax[0].set_title("Original matrix")
for (j, i), number in np.ndenumerate(original):
    ax[0].text(i, j, number, color = 3 * [0.8], ha = "center", va = "center")

# Repeat for submatrix
ax[1].imshow(C, cmap = "gray", origin = "upper", vmin = 0, vmax = np.max(C))
ax[1].set_title("Submatrix")
for (j, i), number in np.ndenumerate(C):
    ax[1].text(i, j, number, color = 3 * [0.8], ha = "center", va = "center")


plt.savefig('Figure_05_06.png',dpi=300)
plt.show()
```

# Exercise 4

```python
A = np.random.randn(16).reshape(4,4)
B = np.random.randn(16).reshape(4,4)

def add_matrices(A, B):
    if A.shape != B.shape:
        raise("Matrices are not the same dimension")
    
    C = np.zeros(A.shape)
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            C[i,j] = A[i,j] + B[i,j]
    
    return C

print(add_matrices(A, B) == (A + B))
```

# Exercise 6

```python
A = np.random.randn(16).reshape(4,4)
B = np.random.randn(16).reshape(4,4)

def multiply_matrices(A, B):
    if A.shape[1] != B.shape[0]:
        raise("Incompatible matrices")
        
    c = np.zeros(A.shape)
    for i in range(B.shape[1]):
        for j in range(A.shape[0]):
            c[j,i] = np.dot(B[:,i], A[j,:])
            
    return c
            
np.isclose(multiply_matrices(A, B), A @ B)
```

# Exercise 11

```python
n = 4
O = np.ones((n,n))
D = np.diag(np.arange(1,5))
S = D @ D

# Pre-multiplying scales rows by diagonal values
pre = S @ O
# Post-multiplying scales columns by diagonal values
pst = O @ S
#  Pre and post by sqrt diagonal (creates multiplication table)
both = D @ O @ D

print(pre, "\n")
print(pst, "\n")
print(both)
```
