```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1

```python
v = np.array([1,2])
w = np.array([4,-6])
vpw = v + w
vmw = v - w
```

```python
plt.figure(figsize = (4, 4))
a1 = plt.arrow(0, 0, v[0], v[1], head_width = 0.3, width = 0.1, color = "k", length_includes_head = True)
a2 = plt.arrow(v[0], v[1], w[0], w[1], head_width = 0.3, width = 0.1, color = [0.5, 0.5, 0.5], length_includes_head = True)
a3 = plt.arrow(0, 0, vpw[0], vpw[1], head_width = 0.3, width = 0.1, color = [0.8, 0.8, 0.8], length_includes_head = True)

plt.grid(linestyle='--',linewidth=.5)
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.legend([a1, a2, a3], ["v", "w", "v + w"])
plt.savefig('Figure_02_02a.png',dpi=300)
plt.title("Vector addition")

plt.show()
```

```python
plt.figure(figsize = (4, 4))
a1 = plt.arrow(0, 0, v[0], v[1], head_width = 0.3, width = 0.1, color = "k", length_includes_head = True)
a2 = plt.arrow(0, 0, w[0], w[1], head_width = 0.3, width = 0.1, color = [0.5, 0.5, 0.5], length_includes_head = True)
a3 = plt.arrow(w[0], w[1], vmw[0], vmw[1], head_width = 0.3, width = 0.1, color = [0.8, 0.8, 0.8], length_includes_head = True)

plt.grid(linestyle='--',linewidth=.5)
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.legend([a1, a2, a3], ["v", "w", "v - w"])
plt.savefig('Figure_02_02b.png',dpi=300)
plt.title("Vector subtraction")

plt.show()
```

# Exercise 2

```python
def l2_norm(v):
    return np.sqrt(np.sum(v*v))

w = np.array([0, 0, 1])
print(l2_norm(w))

w = np.array([1, 2, 3])
print(l2_norm(w), np.linalg.norm(w))
```

# Exercise 3

```python
def unit_vec(v):
    return v / np.linalg.norm(v)

w = np.array([1, 2, 3])
print(unit_vec(w))
print(np.linalg.norm(unit_vec(w)))
print(unit_vec(np.zeros(5)))
```

# Exercise 4

```python
def scale_vec(v, s):
    return (v * s) / np.linalg.norm(v)

w = np.array([1, 2, 3])
print(scale_vec(w, 5))
print(np.linalg.norm(scale_vec(w, 5)))
```

# Exercise 9

```python
np.random.seed(0)
t = np.random.randn(2)
r = np.random.randn(2)

t_par = r * (np.dot(t, r) / np.dot(r, r))
t_per = t - t_par
print(t, t_par + t_per)
```

```python
plt.figure(figsize=(4, 4))

plt.plot([0, t[0]], [0, t[1]], color="k", linewidth = 3,label = "t")
plt.plot([0, r[0]], [0, r[1]], color=[.7, .7, .7],linewidth=3,label = "r")
plt.plot([0, t_par[0]], [0, t_par[1]], 'k--', linewidth = 3, label = "proj")
plt.plot([0, t_per[0]], [0, t_per[1]], 'k:', linewidth = 3, label = "perp")

plt.axis('equal')
plt.legend()
plt.savefig('Figure_02_08.png',dpi=300)
plt.show()
```
