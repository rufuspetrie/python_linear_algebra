# Exercise 1

```python
np.random.seed(0)
v = np.random.randn(10)
x = v + np.random.randn(10)

def cors(v1, v2):
    pearson = np.dot(v1 - v1.mean(), v2 - v2.mean()) / (np.linalg.norm(v1 - v1.mean()) * np.linalg.norm(v2 - v2.mean()))
    cosine_s =  np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    return pearson, cosine_s
    
print(cors(v, x))
print(cors(v - v.mean(), x - x.mean()))
```

- Note: need n in denominator if using standard deviation functions
- Notice how sums of products can be described as dot products and sums of squares as norms
- Notice how the Pearson correlation and cosine similarity are the same when the data is demeaned

# Exercise 2

```python
v = np.arange(4)
offsets = np.arange(-50, 51)
res = np.zeros((len(offsets), 2))
for i in range(len(offsets)):
    res[i,:] = cors(v, v + offsets[i])

plt.figure(figsize = (8, 4))
h = plt.plot(offsets, res)
h[0].set_color("k")
h[0].set_marker("o")
h[1].set_color([0.7, 0.7, 0.7])
h[1].set_marker("s")
plt.xlabel("Offset")
plt.ylabel("r or c")
plt.legend(["Pearson Correlation", "Cosine Similarity"])
plt.savefig('Figure_04_02.png',dpi=300)
plt.show()
```

- Once again, notice that the Pearson correlation remains constant because it demeans data
- On the other hand, notice how Cosine Similarity varies, even where data has a deterministic relationship

# Exercise 4

```python
import time

def cor(v1, v2):
    v1 -= v1.mean()
    v2 -= v2.mean()
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

reps = 1000
size = 500

start = time.time()
for i in range(reps):
    v = np.random.randn(size)
    x = np.random.randn(size)
    cor(v, x)
t1 = time.time() - start

start = time.time()
for i in range(reps):
    v = np.random.randn(size)
    x = np.random.randn(size)
    np.corrcoef(v, x)
t2 = time.time() - start

print("Barebones algorithm: \t%0.4f\nLibrary algorithm: \t%0.4f" % (t1, t2))
```

- Note that you can save a ton of time by demeaning both vectors once and saving the result instead of doing it in the numerator and denominator

# Exercise 8

```python
n_per_clust = 50
blur = 1
A = [1, 1]
B = [-3, 1]
C = [3, 3]

a = [A[0] + np.random.randn(n_per_clust) * blur, A[1] + np.random.randn(n_per_clust) * blur]
b = [B[0] + np.random.randn(n_per_clust) * blur, B[1] + np.random.randn(n_per_clust) * blur]
c = [C[0] + np.random.randn(n_per_clust) * blur, C[1] + np.random.randn(n_per_clust) * blur]
data = np.transpose(np.concatenate((a, b, c), axis = 1))

plt.plot(data[:, 0], data[:, 1], "ko", markerfacecolor = "w")
plt.title("Unclustered data")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
k = 3
ridx = np.random.choice(range(len(data)), k, replace = False)
centroids = data[ridx, :]

fig, ax = plt.subplots(2, 2, figsize = (6, 6))
ax = ax.flatten()
line_colors = [3 * [0], 3 * [0.4], 3 * [0.8]]

ax[0].plot(data[:, 0], data[:, 1], "ko", markerfacecolor = "w")
ax[0].plot(centroids[:, 0], centroids[:, 1], "ko")
ax[0].set_title("Iteration 0")
ax[0].set_xticks([])
ax[0].set_yticks([])

for i in range(3):
    
    # Calculate distance from centroids
    dists = np.zeros((data.shape[0], k))
    for idx in range(k):
        dists[:, idx] = np.sum((data - centroids[idx, :])**2, axis = 1)
    
    # Update clusters
    gidx = np.argmin(dists, axis = 1)
    
    # Recompute centers
    for idx in range(k):
        centroids[idx, :] = [np.mean(data[gidx == idx, 0]), np.mean(data[gidx == idx, 1])]

    for idx in range(len(data)):
        ax[i + 1].plot([data[idx, 0], centroids[gidx[idx], 0]], [data[idx, 1], centroids[gidx[idx], 1]],
                      color = line_colors[gidx[idx]])
    ax[i + 1].plot(centroids[:, 0], centroids[:, 1], "ko")
    ax[i + 1].set_title("Iteration %d" % (i + 1))
    ax[i + 1].set_xticks([])
    ax[i + 1].set_yticks([])

plt.savefig('Figure_04_03.png',dpi=300)
plt.show()
```
