# Exercise 2

- Algorithm:
    - Minors matrix - matrix of determinants found by excluding the (i,j) row and column for each index
    - Grid matrix - same pattern as determinant matrix
    - Cofactors matrix - minors matrix times the grid matrix (Hadamard multiplication, not matrix)
- To find inverse, compute the transpose of the cofactors matrix divided by the determinant of A

```python
n = 4
A = np.random.randn(n, n)
M = np.zeros((n,n))
G = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        rows = [True] * n
        cols = [True] * n
        rows[i], cols[j] = False, False
        M[i,j] = np.linalg.det(A[rows,:][:,cols]) # need double slice to exclude a row then a column
        G[i,j] = (-1)**(i+j)
        
C = M * G
A_inv = C.T / np.linalg.det(A)
A_invl = np.linalg.inv(A)
np.all(np.isclose(A_inv, A_invl))
```

```python
fig, ax = plt.subplots(2, 3, figsize = (14,7))

ax[0,0].imshow(M, cmap = "gray")
ax[0,0].set_title("Minors matrix")
ax[0,1].imshow(G, cmap = "gray")
ax[0,1].set_title("Grid matrix")
ax[0,2].imshow(C, cmap = "gray")
ax[0,2].set_title("Cofactor matrix")
ax[1,0].imshow(A_inv, cmap = "gray")
ax[1,0].set_title("Adjugate inverse")
ax[1,1].imshow(A_invl, cmap = "gray")
ax[1,1].set_title("Library inverse")
ax[1,2].imshow(A @ A_inv, cmap = "gray")
ax[1,2].set_title("A @ A_inv")

for a in ax.flatten():
    a.set_xticks([])
    a.set_yticks([])
plt.savefig('Figure_08_03.png',dpi=300)
plt.show()
```

# Exercise 4

- Wide matrix - $A \in \mathbb{R}^{m \times n}$ with $n > m$
- Right inverse exists for wide matrices because the larger number must cancel when multiplying by a transpose
- $A A^{\prime} \in \mathbb{R}^{m \times m}$ has full rank when formed from randomized matrices, so its inverse exists
- $A A^{\prime} (A A^{\prime})^{-1} = I \implies I_{r} = A^{\prime}(A A^{\prime})^{-1}$

```python
W = np.random.randint(-10, 11, size = (4,40))
I_r = W.T @ np.linalg.inv(W @ W.T)
print(np.round(W @ I_r, 6))
```

```python
fig,ax = plt.subplots(2, 2, figsize=(10,10))

ax[0,0].imshow(W, cmap = "gray")
ax[0,0].set_title("Wide matrix")
ax[0,1].imshow(I_r, cmap = "gray")
ax[0,1].set_title("Right inverse")
ax[1,0].imshow(W @ I_r, cmap = "gray")
ax[1,0].set_title("W @ I_r")
ax[1,1].imshow(I_r @ W, cmap = "gray")
ax[1,1].set_title("I_r @ W")

for a in ax.flatten():
    a.set_xticks([])
    a.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
T = np.array([1, 0.5, 0, 0.5]).reshape(2,2)
T_i = np.linalg.inv(T)
theta = np.linspace(0, 2*np.pi - 2*np.pi/20, 20)
points = np.vstack((np.cos(theta),np.sin(theta)))
points_t = T @ points
points_0 = T_i @ points_t

plt.figure(figsize = (6,6))
plt.plot(points[0,:], points[1,:], "ko", label = "Original")
plt.plot(points_t[0,:], points_t[1,:], "s", color = [0.7]*3, label = "Transformed")
plt.plot(points_0[0,:], points_0[1,:], "rx", label = "Inverted", markersize = 15)
plt.axis("square")
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.legend()
plt.savefig('Figure_08_06.png',dpi=300)
plt.show()
```

# Exercise 9

```python
# Notice that Hilbert matrices are symmetrical, so it's better to
    # implement it with an inverted outer product (done with summation)
def hilbert(k):
    H = np.arange(1,k+1).reshape(1,-1)
    return 1 / (H + H.T - 1)

H = hilbert(10)
H_i = np.linalg.inv(H)
HH_i = H @ H_i

fig,ax = plt.subplots(1, 3, figsize=(12,6))
h = [0,0,0]

# the matrix
h[0] = ax[0].imshow(H, cmap = "gray")
ax[0].set_title("Hilbert matrix")
h[1] = ax[1].imshow(H_i, cmap = "gray")
ax[1].set_title("Hilbert inverse")
h[2] = ax[2].imshow(HH_i, cmap = "gray")
ax[2].set_title("Hilbert times inverse")

for i in range(2):
    fig.colorbar(h[i], ax = ax[i], fraction = 0.045)
    ax[i].set_xticks([])
    ax[i].set_yticks([])

plt.tight_layout()
plt.savefig('Figure_08_05.png',dpi=300)
plt.show()
```

# Exercise 10

```python
sizes  = np.arange(3,13)
errors = np.zeros((len(sizes),2))
conds = np.zeros((len(sizes),2))

for i, k in enumerate(sizes):
    
    H = hilbert(k)
    H_i = np.linalg.inv(H)
    HH_i = H @ H_i
    errors[i,0] = np.sqrt(np.sum((HH_i - np.eye(k))**2))
    conds[i,0] = np.linalg.cond(H)
    
    H = np.random.randn(k,k)
    H_i = np.linalg.inv(H)
    HH_i = H @ H_i
    errors[i,1] = np.sqrt(np.sum((HH_i - np.eye(k))**2))
    conds[i,1] = np.linalg.cond(H)
    
fig, ax = plt.subplots(1, 2, figsize = (14,5))
h = ax[0].plot(sizes, np.log(errors), "s-", markersize = 12)
h[0].set_color("k")
h[0].set_marker("o")
h[1].set_color("gray")
ax[0].legend(["Hilbert", "Random"])
ax[0].set_xlabel("Matrix size")
ax[0].set_ylabel("Log Euclidean distance")
ax[0].set_title("Distance to identity matrix")

h = ax[1].plot(sizes, np.log(conds), "s-", markersize = 12)
h[0].set_color("k")
h[0].set_marker("o")
h[1].set_color("gray")
ax[1].legend(["Hilbert", "Random"])
ax[1].set_xlabel("Matrix size")
ax[1].set_ylabel("Log Kappa")
ax[1].set_title("Matrix condition number")

plt.savefig('Figure_08_07.png',dpi=300)
plt.show()
```

- Note: the condition number of a matrix is the ratio of the largest and smallest singular values
- Because we invert the Hilbert matrix, the distance to the identity matrix should be 0
- Therefore, we can see from the graphs that the Hilbert matrix accumulates a ton of error compared to random matrices as its dimension increases
